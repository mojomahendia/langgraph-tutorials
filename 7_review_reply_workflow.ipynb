```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict, Literal
from dotenv import load_dotenv
from pydantic import BaseModel, Field
```

```python
load_dotenv()
model = ChatOpenAI(model='gpt-4o-mini')
```

```python
# schema for str output from llms
class SentimentSchema(BaseModel):
    sentiment: Literal['positive','negative'] = Field(description='Sentiment of the review')

class DiagnosisSchema(BaseModel):
    issue_type: Literal['UX', 'Performance','Bug', 'Support', 'Other']=Field(description="The category of issue mentioned in the review")
    tone: Literal['angry','frustrated','disappointed','calm',]=Field(description='emotional tone expressed by user.')
    urgency: Literal['low','mid','high']=Field(description='How urgent or critical the issue appear to be.')
```

```python
# creating str models
structured_model = model.with_structured_output(SentimentSchema)
structured_model1 = model.with_structured_output(DiagnosisSchema)
```

```python
prompt = 'The software too ugly'
structured_model.invoke(prompt).sentiment
```
