```python
from langgraph.graph import START,END, StateGraph
from typing import TypedDict, Literal
```

```python
class QuadState(TypedDict):
    a: int
    b: int
    c: int
    equation: str
    dis: float
    result: str
```

```python
def show_equation(state: QuadState):
    equation = f"{state['a']}x2+{state['b']}x+{state['c']}"
    # equation = f'{state["a"]}x2{state["b"]}x{state["c"]}'
    return {'equation':equation}
def calculate_discrimant(state: QuadState):
    dis = state['b']**2-4*state['a']*state['c']
    return{'dis':dis}
```

```python
graph = StateGraph(QuadState)

graph.add_node('show_equation',show_equation)
graph.add_node('calculate_discrimant',calculate_discrimant)

graph.add_edge(START, 'show_equation')
graph.add_edge("show_equation", 'calculate_discrimant')
graph.add_edge('calculate_discrimant', END)

workflow = graph.compile()
workflow

```
